**Задача_1**

**Имеется колода в 52 карты. Найти число возможностей вытянуть из неё 4 карты так, чтобы среди них был хотя бы один туз.**
<hr>

<center>Решение: <br>

Разобъем колоду на две части: в одной тузы (4 карты), в остальной - оставштеся 48 карт. <br>

1) Количество способов вытащить **ОДНОГО** туза => ($C_{4}^1$) и при этом вытащить ТРИ "не туза" => ($C_{48}^3$) <br>
Таких комбинаций: $A_1 = C_{4}^1 \cdot C_{48}^3.$ <br>

2) Количество способов вытащить **ДВУХ** тузов => ($C_{4}^2$) и при этом вытащить ДВА "не туза" => ($C_{48}^2$) <br>
Такое событие наступит с вероятносТаких комбинацийтью : $A_2 = C_{4}^2 \cdot C_{48}^2.$ <br>

3) Количество способов вытащить **ТРЁХ** тузов => ($C_{4}^3$) и при этом вытащить ОДНОГО "не туза" => ($C_{48}^1$) <br>
Таких комбинаций : $A_3 = C_{4}^3 \cdot C_{48}^1.$ <br>

4) Количество способов вытащить **ЧЕТЫРЁХ** тузов => ($C_{4}^4$)
Таких комбинаций : $A_4 = C_{4}^4 \cdot C_{48}^0.$ <br>


А1...А4 комбинации не повторяются, поэтому для итогового значения можем сложить $A_1 ...A_4 $



Для того, чтобы найти искомое число, нужно сложить количество этих событий: $A = A_1 + A_2 + A_3 + A_4$
    
    

Сочетания считаем по формуле: $C_{n}^k = \dfrac{n!}{k! \cdot (n - k)!}$    
    

```python
import numpy as np
import math
```

```python
def get_combination(n, k):
    "сочетание"
    c = np.math.factorial(n) // (np.math.factorial(k) * np.math.factorial(n - k))
    return c
```

```python
a1 = get_combination(4,1) * get_combination(48,3)
a2 = get_combination(4,2) * get_combination(48,2)
a3 = get_combination(4,3) * get_combination(48,1)
a4 = get_combination(4,4) * get_combination(48,0)
a = a1 + a2 + a3 + a4
print(f'ОТВЕТ: число возможностей вытянуть из 52х-карточной колоды 4 карты так, чтобы среди них был хотя бы один туз равно {a}')
```

**Задача_2**

**Из 60 вопросов, входящих в экзаменационные билеты, студент знает 50. Случайным образом студент вытягивает 3 вопроса. Какова вероятность, что все выбранные вопросы знакомы студенту?**
<hr>

<center>Решение: <br>

Наше событие A  - все выбранные вопросы знакомы студенту.<br>
Вероятность найдем по формуле классической вероятности: $P(A) = \dfrac{m}{n} $
    
Где: n — общее число исходов (3 из 60) ($n = C_{60}^3$) , а m — число исходов, которые влекут за собой событие A (3 из 50) ($m = C_{50}^3$).
   
Сочетания считаем по формуле: $C_{n}^k = \dfrac{n!}{k! \cdot (n - k)!}$    

```python
def get_combination(n, k):
    "сочетание"
    c = np.math.factorial(n) // (np.math.factorial(k) * np.math.factorial(n - k))
    return c
```

```python
 p_1 = get_combination(50,3) / get_combination(60,3)
    
print(f'ОТВЕТ: вероятность, что все выбранные вопросы знакомы студенту равна {round(p_1,5)}')
```

**Задача_3**

**Из 60 вопросов, входящих в экзаменационные билеты, студент знает 50. Случайным образом студент вытягивает 3 вопроса. Какова вероятность что два из трёх вопросов знакомы студенту?**
<hr>

<center>Решение: <br>

Наше событие A  - все выбранные вопросы знакомы студенту.<br>
Вероятность найдем по формуле классической вероятности: $P(A) = \dfrac{m}{n} $
    
Где: n — общее число исходов (3 из 60) ($n = C_{60}^3$) , а m — число исходов, которые складываются из двух событий: A1 (2 из 50) ($m = C_{50}^2$) и А2 (1 из 10) ($m = C_{10}^1$), которые перемножаютмя (потому что независимы)
   
Сочетания считаем по формуле: $C_{n}^k = \dfrac{n!}{k! \cdot (n - k)!}$    

```python
A1_2x50 = get_combination(50,2)
A2_1x10 = get_combination(10,1)

p_2 = (A1_2x50 * A2_1x10) / get_combination(60,3)
print(f'ОТВЕТ: вероятность, что два из трёх вопросов знакомы студенту равна {round(p_2,5)}')
```

**Задача_4**

**Допустим, имеется некоторая очень редкая болезнь (поражает 0.1% населения).<br> Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным.<br> Врач говорит вам, что этот тест верно выявляет 99% больных этой болезнью и всего лишь в 1% случаев даёт ложный положительный ответ.<br>
_Вопрос_: какова вероятность, что вы действительно больны ей?<br><br>
_Подсказка_: используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.**
<hr>

<center>Решение: <br>

Событие $A$ - пациент болен ($\overline{A}$ - здоров).<br>
Гипотеза $Н_1$ - тест "+".<br>
    

Тогда вероятность того, что пациент болен найдем по формуле Байеса: $$P(A|Н_1) = \frac{P(Н_1|A) \cdot P(A)}{P(Н_1)}$$
    
    
Где:  <br>
$P(A|Н_1)$ - вероятность того, что при положительном тесте пациент болен<br>
$P(A)$ - вероятность болезни (у населения)<br>
$P(Н_1|A)$ - вероятность положительного теста при болезни <br>
$P(Н_1)$   - полная вероятность положительного теста <br>
    
    
$P(Н_1)$   - полная вероятность положительного теста находится по формуле полной вероятности: <br>
    $$P(Н_1) = P(Н_1| A) \cdot P(A) + P \left( Н_1 | \overline{A} \right) \cdot P \left( \overline{A} \right)$$
    

По условию задачи: <br>
$P(A) = 0,001$  <br>
$P(Н_1|A) = 0,99$ <br>
$P \left( Н_1 | \overline{A} \right) = 0,01$ - вероятность ошибки теста<br>
$P \left( \overline{A} \right) = 0,999 $ - вероятность отсутствия болезни у населения
    


```python
p = 0.99 * 0.001 / (0.99 * 0.001 + 0.01 * 0.999)
print(f'ОТВЕТ: вероятность, того, что пациент действительно болен равна {round(p,5)}')
```
